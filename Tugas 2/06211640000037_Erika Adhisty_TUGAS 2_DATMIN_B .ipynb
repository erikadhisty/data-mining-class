```python
#imputasi missing value data numerik
data['Usia'] = data['Usia'].fillna((data['Usia'].mean())) 
data['Tekanan Darah'] = data['Tekanan Darah'].fillna((data['Tekanan Darah'].median()))
data.isnull().any()
```

```python
# imputasi untuk data numerik yang lain
data['Nafsu Makan']=data['Nafsu Makan'].fillna(99)
data['Pedal Edema']=data['Pedal Edema'].fillna(111)
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus
data['Eritrosit'] = data.fillna(data['Eritrosit'].value_counts().index[0])
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Hipertensi'] = LabelEncoder().fit_transform(data['Hipertensi'])
data['Diabetes Melitus'] = LabelEncoder().fit_transform(data['Diabetes Melitus'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Usia'] = StandardScaler().fit(data[['Usia']]).transform(data[['Usia']])
data['Packed Cell'] = StandardScaler().fit(data[['Packed Cell']]).transform(data[['Packed Cell']])
data.head()
```

## Normalize

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['Usia'] = Normalizer().fit(data[['Usia']]).transform(data[['Usia']])
data['Packed Cell'] = Normalizer().fit(data[['Packed Cell']]).transform(data[['Packed Cell']])
data.head()
```
