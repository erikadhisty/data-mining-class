```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
data=pd.read_csv('D:\online_retail.csv')
data.head()
```

```python
np.sum(data.isnull())
```

```python
for i, j in data[data['Description'].isnull()].iterrows(): data['Description'][i]="Code-"+str(j['StockCode'])
```

```python
np.sum(data.isnull())
```

```python
x=data.groupby('InvoiceNo')
beli=[]
for i,j in x:
    beli.append(list(j['Description'].map(str)))
```

```python
data1=pd.DataFrame(beli)
data1
```

```python
association_rules=apriori(beli,min_support=0.0045,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
    pair = item[0]
    items = [x for x in pair]
    print("rule: " + items[0] + "->" + items[1])
    
    print("support: " + str(item[1]))
    
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=====================================")
```
