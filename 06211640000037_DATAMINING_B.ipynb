```python
import pandas as pd
import numpy as np
```

```python
#Merging
provinsi = pd.DataFrame({'id': ['id0', 'id1', 'id2', 'id3', 'id4'],
                     'Provinsi': ['Jawa Timur', 'Jawa Barat', 'Kalimantan Timur', 'Sumatera Selatan', 'Sulawesi Tengah'],
                     'Ibu Kota': ['Surabaya', 'Bandung', 'Samarinda', 'Palembang', 'Palu']})

karakteristik = pd.DataFrame({'id': ['id0', 'id1', 'id2', 'id3', 'id4'],
                      'Luas': ['47.922 km²', '35.222,18 km²', '127.346,92 km²', '91.592,43 km²', '61.841,29 km²'],
                      'Zona Waktu': ['WIB', 'WIB', 'WITA', 'WIB', 'WITA']})

result1 = pd.merge(provinsi, karakteristik, on='id')
result1
```

```python
#Multiple key
provinsi = pd.DataFrame({'id': ['id0', 'id0', 'id1', 'id1'],
                        'Singkatan': ['Jatim', 'Jabar', 'Kaltim', 'Sumsel'],
                        'Provinsi': ['Jawa Timur', 'Jawa Barat', 'Kalimantan Timur', 'Sumatera Selatan'],
                        'Ibu Kota': ['Surabaya', 'Bandung', 'Samarinda', 'Palembang']})

karakteristik = pd.DataFrame({'id': ['id0', 'id0', 'id1', 'id1'],
                      'Singkatan': ['Jatim', 'Jabar', 'Kaltim', 'Sumsel'],
                      'Luas': ['47.922 km²', '35.222,18 km²', '127.346,92 km²', '91.592,43 km²'],
                      'Zona Waktu': ['WIB', 'WIB', 'WITA', 'WIB']})

result3 = pd.merge(provinsi, karakteristik, on=['id', 'Singkatan'])
result3
```

```python
#Merge methods (default how="inner")
result_left = pd.merge(provinsi, karakteristik, how='left', on=['id', 'Singkatan'])
result_left
```

```python
#Merge methods (default how="inner")
result_right = pd.merge(provinsi, karakteristik, how='right', on=['id', 'Singkatan'])
result_right
```

```python
result_outer = pd.merge(provinsi, karakteristik, how='outer', on=['id', 'Singkatan'])
result_outer
```

```python
result_inner = pd.merge(provinsi, karakteristik, how='inner', on=['id', 'Singkatan'])
result_inner
```

```python
##Joining 
provinsi = pd.DataFrame({'Provinsi': ['Jawa Timur', 'Jawa Barat', 'Kalimantan Timur'],
                        'Ibu Kota': ['Surabaya', 'Bandung', 'Samarinda']},
                    index=['id0', 'id1', 'id2'])
karakteristik = pd.DataFrame({'Singkatan': ['Jatim', 'Jabar', 'Kaltim'],
                      'Luas': ['47.922 km²', '35.222,18 km²', '127.346,92 km²']},
                    index=['id0', 'id2', 'id3'])
result = provinsi.join(karakteristik) #Based Provinsi ditambahkan karakteristik
result
```

```python
result = karakteristik.join(provinsi, how='outer') #Seluruh data masuk
result
```

```python
result = karakteristik.join(provinsi, how='inner') # Hanya data yang lengkap
result
```

```python
#The data alignment here is on the indexes (row labels)
result = pd.merge(provinsi, karakteristik, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(provinsi, karakteristik, left_index=True, right_index=True, how='inner')
result
```

```python
provinsi = pd.DataFrame({'Provinsi': ['Jawa Timur', 'Jawa Barat', 'Kalimantan Timur', 'Sumatera Selatan'],
                        'Ibu Kota': ['Surabaya', 'Bandung', 'Samarinda', 'Palembang'],
                   'key': ['id0', 'id1', 'id0', 'id1']})
karakteristik = pd.DataFrame({'Luas': ['47.922 km²', '35.222,18 km²'],
                      'Zona Waktu': ['WIB', 'WIB']},
                      index=['id0', 'id1'])
result = provinsi.join(karakteristik, on='key')
result
```

```python
#Concatenating
fc1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
fc2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
fc3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [fc1, fc2, fc3]
result = pd.concat(frames)
result
```

```python
#Add category
result = pd.concat(frames, keys=['K', 'L', 'M'])
result
```

```python
#Default join=outer (join all rows)
fc4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

result = pd.concat([fc1, fc4], axis=0) #axis=1 is for concated by y-axis; default axis=0
result
```

```python
#Join = inner (join only rows that exist in both datasets) 
result = pd.concat([fc1, fc4], axis=1, join='inner')
result
```

```python
#Join the exact rows in df1 
result = pd.concat([fc1, fc4], axis=1, join_axes=[fc1.index])
result
```

```python
## Concatenating using append
result = fc1.append(fc2) #default axis=0
result
```

```python
result = fc1.append(fc4)
result
```

```python
result = fc1.append([fc2, fc3]) #multiple datasets to concatenate
result
```

```python
# For DataFrames which don’t have a meaningful index, 
result = pd.concat([fc1, fc4], ignore_index=True)
result
```

```python
result = fc1.append(fc4, ignore_index=True)
result
```

```python
# Appending rows to a DataFrame
se7 = pd.Series(['X0', 'X1', 'X2', 'X3'], index=['A', 'B', 'C', 'D'])
result = fc1.append(se7, ignore_index=True)
result
```
